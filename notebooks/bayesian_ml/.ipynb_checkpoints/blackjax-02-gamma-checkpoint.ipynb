```python
import distrax
import matplotlib.pyplot as plt
import seaborn as sns
import jax
jax.config.update("jax_enable_x64", True)
import numpy as np
import jax.numpy as jnp
import jax.random as random
import blackjax
```

```python
dist = distrax.Gamma(3, 1)
```

```python
samples = dist.sample(seed=2, sample_shape=100)
```

```python
plt.hist(samples)
```

```python
sns.kdeplot(samples, shade=True, bw_adjust=2)
```

```python
def logprob_fn(theta):
    log_conc, log_rate = theta
    conc = jnp.exp(log_conc)
    rate = jnp.exp(log_rate)
    dist = distrax.Gamma(conc, rate)
    return dist.log_prob(samples).sum()
```

```python
logprob_fn((jnp.log(3), jnp.log(1)))
```

```python
rng_key = random.PRNGKey(314)
M = 2
theta0 = random.multivariate_normal(rng_key, 0.1 + jnp.zeros(M), jnp.eye(M))

rmh = blackjax.rmh(logprob_fn, sigma=jnp.ones(M) * 0.1)
initial_state = rmh.init(theta0)
```

```python
def inference_loop(rng_key, kernel, initial_state, num_samples):
    @jax.jit
    def one_step(state, rng_key):
        state, _ = kernel(rng_key, state)
        return state, state

    keys = jax.random.split(rng_key, num_samples)
    _, states = jax.lax.scan(one_step, initial_state, keys)

    return states
```

```python
_, rng_key = random.split(rng_key)
states = inference_loop(rng_key, rmh.step, initial_state, 5_000)
```

```python
burnin = 200
```

```python
plt.plot(states.position[burnin:, 0], label='log_conc')
plt.plot(states.position[burnin:, 1], label='log_rate')
plt.legend()
```

```python
log_conc_hat = states.position[burnin:, 0].mean()
conc_hat = jnp.exp(log_conc_hat)
conc_hat
```

```python
log_rate_hat = states.position[burnin:, 1].mean()
rate_hat = jnp.exp(log_rate_hat)
rate_hat
```
